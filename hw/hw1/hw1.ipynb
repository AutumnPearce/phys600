```python
# imports
from matplotlib import pyplot as plt
import numpy as np
import scipy.integrate as integrate
```

The analytic solution to the integral can be calculated for $\Omega=0$ and $\Omega=1$ through basic simplifications.

When $\Omega=0$, the integral becomes:  $4(1+z)^{1\over4}-4$

When $\Omega=1$, the integral becomes:  $-2(1+z)^{-{1\over2}}+2$

```python
def func(z, Omega):
    def f(z):
        return np.sqrt(Omega*(1+z)**3 + (1-Omega)*(1+z)**1.5) ** -1
    
    return integrate.quad(f, 0, z)[0]

#Omega = 0, analytic:
def analytic0(z):
    return 4*(1+z)**.25 - 4

#Omega = 1, analytic:
def analytic1(z):
    return -2*(1+z)**-.5 + 2
```

```python
z=np.linspace(0,1,200)
all_Omegas = [0, .3, .7, 1]
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
fig.set_facecolor('lightgray')
ax.set_title(r"$\int\frac{dz}{[\Omega(1+z)^3+(1-\Omega)(1+z)^{\frac{3}{2}}]^{\frac{1}{2}}}$", fontsize=20)
ax.set_xlabel('z', fontsize=15)
ax.set_ylabel('Value', fontsize=15)

colors = ['firebrick', 'magenta', 'turquoise', 'teal']

#non-analytic
for i,Omega in enumerate(all_Omegas):
    vals = [func(z,Omega) for z in z]
    ax.plot(z, vals, label=str(Omega), color=colors[i], linewidth = 2)

#Omega=0 Analytic
vals = [analytic0(z) for z in z]
ax.plot(z, vals, label='0, analytic', color='black', linewidth = 4, linestyle=(0,(1,5)))

#Omega=1 Analytic
vals = [analytic1(z) for z in z]
ax.plot(z, vals, label='1, analytic', color='black', linewidth = 4, linestyle=(0,(3,3,1,3)))

ax.legend(title='Omega Value',handlelength=3.2,title_fontsize=11.5)
```
